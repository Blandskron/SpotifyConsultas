```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/drive/MyDrive/top_artists.csv')
df
```

```python
# Imprimir las primeras filas del archivo
print(df.head())

# Obtener información general del archivo
print(df.info())

# Obtener estadísticas descriptivas de las columnas numéricas
print(df.describe())

# Contar la cantidad de artistas por género musical
print(df['Género musical'].value_counts())

# Obtener la cantidad promedio de seguidores por género musical
print(df.groupby('Género musical')['Seguidores en Spotify'].mean())
```

```python
# Gráfico de barras de la cantidad de artistas por género musical
df['Género musical'].value_counts().plot(kind='bar')
plt.title('Cantidad de artistas por género musical')
plt.xlabel('Género musical')
plt.ylabel('Cantidad de artistas')
plt.show()

# Gráfico de caja de la cantidad de seguidores por género musical
df.boxplot(column='Seguidores en Spotify', by='Género musical')
plt.title('Cantidad de seguidores por género musical')
plt.xlabel('Género musical')
plt.ylabel('Seguidores en Spotify')
plt.show()
```

```python
# Agrupar los datos por el género musical
artistas_por_genero = df.groupby('Género musical')

# Imprimir la cantidad de artistas por género musical
print(artistas_por_genero.size())

# Imprimir los nombres de los artistas por género musical
for nombre, grupo in artistas_por_genero:
    print(nombre)
    print(grupo['Nombre del artista'])
```

```python
# Filtrar las filas correspondientes al género Reggaeton
artistas_reggaeton = df.loc[df['Género musical'] == 'Reggaeton']

# Imprimir los nombres de los artistas y sus números de seguidores
print(artistas_reggaeton[['Nombre del artista', 'Seguidores en Spotify']])
```

```python
# Filtrar las filas correspondientes al género Reggaeton
artistas_reggaeton = df.loc[df['Género musical'] == 'Reggaeton']

# Seleccionar los 10 artistas con más seguidores
top_10_reggaeton = artistas_reggaeton.nlargest(10, 'Seguidores en Spotify')

# Crear un gráfico de barras horizontales
plt.barh(top_10_reggaeton['Nombre del artista'], top_10_reggaeton['Seguidores en Spotify'])
plt.title('Top 10 artistas de Reggaeton con más seguidores')
plt.xlabel('Seguidores en Spotify')
plt.ylabel('Artista')
plt.show()





```
